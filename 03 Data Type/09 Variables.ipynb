  <h2>Variable Names</h2>
  <div class="container">
    <div class="card">
      <p>Variable names can be short like <code>m</code> or <code>n</code>, but longer names are often better — they make code more readable by clearly describing the variable’s purpose.</p>
      <strong>In Python, variable names can:</strong>
      <ul>
      <li>Be as long as you want.</li>
      <li>Use both small and capital letters (a-z, A-Z).</li>
      <li>Have numbers (0-9), but not at the start.</li>
      <li>Use the underscore (_) character.</li>
    </ul>
    <strong>There are different ways to write long name:</strong>
    <ul>
      <li><strong>Camel Case</strong>: First letter of each word (except first) is uppercase — e.g., <code>numberOfCollegeGraduates</code>.</li>
      <li><strong>Pascal Case</strong>: First letter of every word is uppercase — e.g., <code>NumberOfCollegeGraduates</code>.</li>
      <li><strong>Snake Case</strong>: Words separated by underscores — e.g., <code>number_of_college_graduates</code>.</li>
    </ul>
    <strong>Key naming rules:</strong>
    <ul>
    <li>Use <strong>snake_case</strong> for variables and functions — e.g., <code>my_variable</code>, <code>calculate_total()</code>.</li>
    <li>Use <strong>PascalCase</strong> for class names — e.g., <code>MyClass</code>, <code>HttpRequestHandler</code>.</li>
    </ul>
    </div>
</div>

```python
name = "Amin"
Age = 20
```

```python
name, Age
```

But this one isn't good because it starts with a number:

```python
1_item = 10
```

Remember, `age` and `Age` are different because Python cares about small and capital letters. The underscore (`_`) is also important. So, all these are different variables:

```python
age = 1
Age = 2
aGe = 3
AGE = 4
a_g_e = 5
_age = 6
age_ = 7
_AGE_ = 8
```

```python
number_of_college_graduates = 2500
```

  <h2>PEP 8</h2>
  <div class="container">
    <div class="card red">
      <p>PEP 8 is Python’s official style guide — a set of conventions to make code readable, consistent, and easy for others to understand and maintain.</p>

<p>Key areas it covers:</p>
<ul>
  <li><strong>Code formatting:</strong> indentation (4 spaces), line length (79 chars), spacing around operators.</li>
  <li><strong>Naming:</strong> use <code>snake_case</code> for variables/functions, <code>PascalCase</code> for classes.</li>
  <li><strong>Code structure:</strong> imports grouping, blank lines, comments, and docstrings.</li>
</ul>

<p>Following PEP 8 is not mandatory, but strongly recommended — especially in teams — to ensure uniform style and smoother collaboration.</p>

<p>Full guide: <a href="https://peps.python.org/pep-0008/">PEP 8 – Style Guide for Python Code</a>.</p>
