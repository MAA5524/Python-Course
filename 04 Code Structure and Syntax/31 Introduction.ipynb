<div class="section-card" id="toc1_">
  <h2>Python Syntax Overview</h2>
  <div class="container">
    <div class="card orange">
      <img src="https://usemynotes.com/wp-content/uploads/2021/05/python-syntax.jpg" alt="Python Syntax Visual">
    </div>
    <div class="card orange">
      <p>Python's syntax defines how a program is written. Unlike many languages, it uses indentation to delimit code blocks, which enforces readability. Python is visually less cluttered, as it avoids semicolons to end lines or excessive parentheses to define blocks.</p>
    </div>
    <div class="card orange" id="toc1_1_">
      <h3>Simple Statements</h3>
      <p>A simple statement in Python is typically a single logical line, such as an assignment or a function call.</p>
      <pre><code>x = 10
print(x)</code></pre>
    </div>
    <div class="card orange" id="toc1_2_">
      <h3>Compound Statements</h3>
      <p>Compound statements (e.g., <code>if</code>, <code>for</code>, <code>while</code>) contain groups of other statements. They consist of a header line ending in a colon and a suite of indented statements.</p>
      <pre><code>if x > 5:
    print("x is greater than 5")</code></pre>
    </div>
  </div>
</div>

  <h2>Code Blocks and Indentation</h2>
  <div class="container">
    <div class="card purple">
      <img src="../images/python-indentation.png" alt="Python Indentation Visual">
    </div>
    <div class="card purple">
      <p style="margin-bottom:0;">A unique feature of Python is that code blocks are defined by their indentation. All lines within a block must be indented by the same amount. The block begins where the indentation increases and ends where it decreases, unlike other languages that use curly braces or keywords.</p>
