<!-- Section 1: What Is a Package Manager -->
<div class="section-card" id="toc1_">
    <h2>What Is a Package Manager</h2>
    <div class="container">
        <div class="card image-card orange">
            <h3>Overview</h3>
            <p>A package manager is a tool that automates the process of installing, upgrading, configuring, and removing computer programs or libraries. In Python, it helps you manage external code written by others.</</p>
            <img src="../images/packege.jpg" alt="Package Manager Overview" style="max-width: 50vw; height: auto; display: block; margin: 10px auto; border-radius: 5px;">
        </div>
        <div class="card orange">
            <h3>Why Do We Need It?</h3>
            <ul>
                <li><strong>Dependency Resolution:</strong> Automatically installs other libraries required by your main package.</li>
                <li><strong>Version Control:</strong> Ensures you use the correct version of a library for your project.</li>
                <li><strong>Ease of Use:</strong> One command replaces hours of manual searching and compiling.</li>
            </ul>
            <pre><code>pip install pandas</code></pre>
        </div>
        <div class="card orange">
            <h3>Popular Package Managers</h3>
            <p>Examples include:</p>
            <ul>
                <li><strong>pip:</strong> Pythonâ€™s default package manager, perfect for installing libraries from PyPI.</li>
                <li><strong>Conda:</strong> A powerful manager for Python and non-Python libraries, ideal for data science.</li>
            </ul>
        </div>
    </div>
</div>

<!-- Section 2: How Package Managers Work (Previously Provided) -->
<div class="section-card" id="toc2_">
    <h2>How Package Managers Work</h2>
    <div class="container">
        <div class="card image-card purple">
            <h3>Overview</h3>
            <p>A package manager acts like a personal assistant for your coding projects, finding, installing, and organizing the tools (libraries) you need from online repositories.</p>
            <img src="../images/how-package-manager-work.png" alt="Package Manager Workflow" style="max-width: 50vw; height: auto; display: block; margin: 10px auto; border-radius: 5px;">
        </div>
        <div class="card purple">
            <h3>Steps of Package Management</h3>
            <p>Package managers follow a clear process to manage libraries:</p>
            <ul>
                <li><strong>Search:</strong> They search repositories (like PyPI or Conda Forge) to find the requested package.</li>
                <li><strong>Download:</strong> They download the package and its dependencies.</li>
                <li><strong>Install:</strong> They install the package, ensuring compatibility with your system.</li>
                <li><strong>Manage Dependencies:</strong> They resolve conflicts by selecting compatible versions.</li>
            </ul>
        </div>
    </div>
</div>

<!-- Section 3: Benefits of Using a Package Manager -->
<div class="section-card" id="toc3_">
    <h2>Benefits of Using a Package Manager</h2>
    <div class="container">
        <div class="card image-card green">
            <h3>Why Use a Package Manager?</h3>
            <p>Package managers simplify coding by automating library management, letting you focus on building projects like a web app or data analysis tool.</p>
            <img src="https://geekflare.com/wp-content/uploads/2023/09/7-package-manager-functions.png" alt="Benefits of Package Manager" style="max-width: 50vw; height: auto; display: block; margin: 10px auto; border-radius: 5px;">
        </div>
        <div class="card green">
            <h3>Key Benefits</h3>
            <ul>
                <li><strong>Saves Time:</strong> Installs libraries quickly with one command, like <code>pip install django</code> for a web app.</li>
                <li><strong>Prevents Conflicts:</strong> Uses virtual environments to keep project tools separate, avoiding version mismatches.</li>
                <li><strong>Ensures Consistency:</strong> Records exact package versions, so your project works the same on any system.</li>
                <li><strong>Simplifies Updates:</strong> Easily updates libraries to their latest versions with commands like <code>pip install --upgrade pandas</code>.</li>
            </ul>
        </div>
    </div>
</div>

<!-- Section 4: Package Repository -->
<div class="section-card" id="toc4_">
    <h2>Package Repository</h2>
    <div class="container">
        <div class="card image-card orange">
            <h3>What Is a Package Repository?</h3>
            <p>A package repository is like an online library where developers share and download code packages, making it easy to access tools for your projects.</p>
            <img src="../images/pypi-pip.png" alt="Package Repository" style="max-width: 50vw; height: auto; display: block; margin: 10px auto; border-radius: 5px;">
        </div>
        <div class="card orange">
            <h3>Popular Repositories</h3>
            <ul>
                <li><strong>PyPI:</strong> The Python Package Index, hosting over 700,000 Python packages like Pandas and Django.</li>
                <li><strong>Conda Forge:</strong> A community-driven repository with pre-compiled packages for Python and non-Python tools, ideal for data science.</li>
            </ul>
            <div class="note">
                <strong>Note:</strong> Repositories like PyPI are like a digital marketplace where you can find tools for any Python project, from web apps to AI.
            </div>
        </div>
        <div class="card orange">
            <h3>Using Mirrors for Stability</h3>
            <p>In regions like Iran, direct access to PyPI can sometimes be slow or restricted. <strong>Mirrors</strong> are local copies of PyPI that provide much faster download speeds and better stability.</p>
            <p>To install a package using a mirror:</p>
            <pre><code>pip install library_name -i https://mirror-pypi.runflare.com/simple</code></pre>
        </div>
    </div>
</div>

<!-- Section 5: Conda vs. Pip -->
<div class="section-card" id="toc5_">
    <h2>Conda vs. Pip</h2>
    <div class="container">
        <div class="card image-card red">
            <p>Pip and Conda are popular package managers, but they serve different needs. Pip is great for Python libraries, while Conda excels in complex projects with non-Python dependencies.</p>
            <img src="https://miro.medium.com/v2/resize:fit:1200/1*cD6_7-slJ-fqo1evJTQbMw.png" alt="Conda vs Pip" style="max-width: 60vw; height: auto; display: block; margin: 10px auto; border-radius: 5px;">
        </div>
        <div class="card red">
            <h3>Key Differences</h3>
            <table>
                <tr>
                    <th>Feature</th>
                    <th>Pip</th>
                    <th>Conda</th>
                </tr>
                <tr>
                    <td><strong>Purpose</strong></td>
                    <td>Installs Python packages from PyPI</td>
                    <td>Manages Python and non-Python packages, plus environments</td>
                </tr>
                <tr>
                    <td><strong>Virtual Environments</strong></td>
                    <td>Requires tools like virtualenv for environments</td>
                    <td>Built-in environment management</td>
                </tr>
                <tr>
                    <td><strong>Binary Packages</strong></td>
                    <td>Some packages need compilation</td>
                    <td>Provides pre-compiled binaries</td>
                </tr>
                <tr>
                    <td><strong>Dependency Management</strong></td>
                    <td>Basic dependency resolution</td>
                    <td>Advanced resolution for complex dependencies</td>
                </tr>
                <tr>
                    <td><strong>Use Case</strong></td>
                    <td>Ideal for simple Python projects (e.g., web apps)</td>
                    <td>Best for data science and scientific computing</td>
                </tr>
            </table>
        </div>
    </div>
</div>
<div class="section-card" id="toc6_">
    <h2>Virtual Environments (venv)</h2>
    <div class="container">
        <div class="card green">
            <h3>The Concept of Isolation</h3>
            <p>Imagine one project needs <code>Django 3.0</code> and another needs <code>Django 4.0</code>. Installing them globally will break your system. A <strong>Virtual Environment</strong> creates a separate "room" for each project.</p>
        </div>
        <div class="card green">
            <h3>Standard Workflow</h3>
            <p>Follow these steps to create and use an environment in your terminal:</p>
            <ul>
                <li><strong>Create:</strong> <code>python -m venv .venv</code></li>
                <li><strong>Activate (Windows):</strong> <code>.venv\Scripts\activate</code></li>
                <li><strong>Activate (Mac/Linux):</strong> <code>source .venv/bin/activate</code></li>
            </ul>
            <div class="note">
                <strong>Pro Tip:</strong> Always name your environment folder <code>.venv</code>. Modern IDEs like VS Code will detect it automatically!
            </div>
        </div>
        <div class="card green">
            <h3>Why we prefer Pip + venv for this course</h3>
            <p>Conda is powerful for Data Science but can be frustrating due to network timeout issues. By using <b>Pip</b> and <b>venv</b>, you gain full control over your environment with high speed using domestic mirrors.</p>
        </div>
    </div>
</div>
