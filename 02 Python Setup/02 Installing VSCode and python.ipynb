<div class="section-card" id="toc1_">
  <h2>Step 1: Installing Python</h2>
  <div class="container">
    <div class="card green" id="toc1_1_">
      <h3>Windows Installation</h3>
      <p>Windows does not come with Python pre-installed. You must install it manually from the official source.</p>
      <ul>
        <li>Go to the official site: <a href="https://www.python.org/downloads/" target="_blank" style="color:#66bb6a;">python.org/downloads</a> and download the latest Windows installer.</li>
        <li>Double-click the downloaded <code>.exe</code> file to open the setup wizard.</li>
      </ul>
      <div class="note-danger">
        <strong>‚ö†Ô∏è CRITICAL STEP:</strong> At the very bottom of the first setup window, you MUST check the box that says <strong>"Add Python to PATH"</strong> before clicking Install Now. If you miss this, your terminal won't recognize Python!
      </div>
    </div>
    <div class="card green" id="toc1_2_">
      <h3>macOS Installation</h3>
      <ul>
        <li>Modern Macs do not come with a supported version of Python.</li>
        <li>Go to <a href="https://www.python.org/downloads/macos/" target="_blank" style="color:#66bb6a;">python.org/downloads/macos</a>.</li>
        <li>Download the <strong>macOS 64-bit universal2 installer</strong>.</li>
        <li>Open the <code>.pkg</code> file and follow the standard installation prompts (Next > Agree > Install).</li>
      </ul>
    </div>
    <div class="card green" id="toc1_3_">
      <h3>Linux Installation</h3>
      <p>Most Linux distributions come with Python pre-installed. However, you might want to install the latest version or specific development tools. Open your terminal and run:</p>
      <ul>
        <li><strong>Ubuntu / Debian:</strong> <code>sudo apt update && sudo apt install python3 python3-pip python3-venv</code></li>
        <li><strong>Fedora:</strong> <code>sudo dnf install python3</code></li>
        <li><strong>Arch Linux:</strong> <code>sudo pacman -S python</code></li>
      </ul>
      <p>To check if Python is installed successfully, open your terminal and type <code>python3 --version</code> or <code>python --version</code>.</p>
    </div>
  </div>
</div>

<!-- Section 2: Installing VS Code -->
<div class="section-card" id="toc2_">
  <h2>Step 2: Installing VS Code</h2>
  <div class="container">
    <div class="card orange">
      <h3>Windows</h3>
      <ul>
        <li>Download the installer from <a href="https://code.visualstudio.com/Download" target="_blank" style="color:#ffa726;">code.visualstudio.com</a>.</li>
        <li>Run the setup. When prompted with "Select Additional Tasks", check the boxes for <strong>"Add to PATH"</strong> and <strong>"Add 'Open with Code' action to Windows Explorer"</strong> (Very useful!).</li>
      </ul>
    </div>
    <div class="card orange">
      <h3>macOS</h3>
      <ul>
        <li>Download the <code>.zip</code> file for macOS from the official site.</li>
        <li>Double-click the downloaded file to extract the <code>Visual Studio Code.app</code>.</li>
        <li><strong>Important:</strong> Drag the app into your <strong>Applications</strong> folder so it shows up in Launchpad.</li>
      </ul>
    </div>
    <div class="card orange">
      <h3>Linux</h3>
      <ul>
        <li>For Debian/Ubuntu, download the <code>.deb</code> file. Install it via terminal: <code>sudo apt install ./&lt;filename&gt;.deb</code></li>
        <li>Or install via Snap store if your distribution supports it.</li>
      </ul>
    </div>

<!-- Section 3: Launching VS Code -->
<div class="section-card" id="toc3_">
  <h2>Step 3: How to Launch Projects Properly</h2>
  <div class="container">
    <div class="card purple">
      <h3>The Golden Rule of VS Code</h3>
      <p>Never open single files in VS Code. Always open <strong>Folders (Workspaces)</strong>. This ensures your project settings, virtual environments, and file paths work correctly.</p>
    </div>
    <div class="card purple">
      <h3>Method 1: From the Terminal (Recommended)</h3>
      <p>If you have added VS Code to your PATH during installation, this is the fastest way to start coding:</p>
      <ol>
        <li>Open your Terminal / Command Prompt.</li>
        <li>Navigate to your project folder using <code>cd</code> (e.g., <code>cd Documents/MyPythonProject</code>).</li>
        <li>Type <code>code .</code> (Note the space and the dot) and press Enter.</li>
      </ol>
      <div class="note">
        <strong>What does <code>code .</code> mean?</strong><br>
        <code>code</code> opens the editor. The <code>.</code> tells it to open the "current directory" as the active workspace.
      </div>
    </div>
    <div class="card purple">
      <h3>Method 2: From the GUI</h3>
      <ul>
        <li><strong>Windows:</strong> Right-click on your project folder and select <strong>"Open with Code"</strong>.</li>
        <li><strong>All OS:</strong> Open VS Code, go to <code>File</code> > <code>Open Folder...</code> and select your project directory.</li>
      </ul>
    </div>
  </div>
