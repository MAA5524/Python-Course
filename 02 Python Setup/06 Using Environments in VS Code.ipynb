<!-- Section 1: Selecting the Interpreter -->
<div class="section-card" id="toc1_">
  <h2>1. Selecting the Python Interpreter</h2>
  <div class="container">
    <div class="card purple">
      <h3>Why do this?</h3>
      <p>VS Code needs to know which Python engine (and which installed packages) to use for your project. This enables correct auto-completion, linting, and execution.</p>
      <h3 style="margin-top: 15px;">Steps (Works for both venv & Conda):</h3>
      <ul>
        <li>Open your project folder in VS Code.</li>
        <li>Press <code>Ctrl + Shift + P</code> (Windows/Linux) or <code>Cmd + Shift + P</code> (Mac) to open the Command Palette.</li>
        <li>Type <strong>Python: Select Interpreter</strong> and hit Enter.</li>
        <li><strong>For venv:</strong> Look for the path ending in <code>.venv/Scripts/python.exe</code></li>
        <li><strong>For Conda:</strong> Look for your environment's name (e.g., <code>my_env</code>) marked with "conda".</li>
      </ul>
      <div class="image-grid">
        <div>
            <p style="text-align: center; color: #aaa; font-size: 0.85rem;">Step 1: Open Command Palette</p>
            <img src="../images/1-show-python-interpreter-list.png" alt="Command Palette">
        </div>
        <div>
            <p style="text-align: center; color: #aaa; font-size: 0.85rem;">Step 2: Select your environment</p>
            <img src="../images/2-select-python-interpreter.png" alt="Select Interpreter">
        </div>
      </div>
      <div class="note">
        <strong>Pro Tip:</strong> You can also just click the Python version number at the very bottom-right corner of the VS Code status bar to quickly open this menu!
      </div>
    </div>
  </div>
</div>

<!-- Section 2: Terminal Activation -->
<div class="section-card" id="toc2_">
  <h2>2. The Integrated Terminal</h2>
  <div class="container">
    <div class="card red">
      <h3>Auto-Activation Magic</h3>
      <p>Once you select your interpreter (Step 1), VS Code becomes smart. When you open a new integrated terminal (<code>Ctrl + `</code>), VS Code will automatically activate your environment!</p>
      <h3 style="margin-top: 15px;">Manual Activation (If Auto fails):</h3>
      <p>If you don't see the environment name in parentheses `(my_env)` or `(.venv)`, you can activate it manually:</p>
      <ul>
        <li><strong>venv (Windows):</strong> <code>.venv\Scripts\activate</code></li>
        <li><strong>venv (Mac/Linux):</strong> <code>source .venv/bin/activate</code></li>
        <li><strong>Conda:</strong> <code>conda activate my_env</code></li>
      </ul>
      <div class="image-grid">
        <div>
            <p style="text-align: center; color: #aaa; font-size: 0.85rem;">activate venv manually</p>

        <div>
            <p style="text-align: center; color: #aaa; font-size: 0.85rem;">activate conda manually</p>
            <img src="../images/vscode-terminal-env-conda.png" alt="Select Environment">
        </div>

<!-- Section 3: Jupyter in VS Code -->
<div class="section-card" id="toc3_">
  <h2>3. Using Jupyter Notebooks</h2>
  <div class="container">
    <div class="card green">
      <h3>Connecting Environments to Notebooks</h3>
      <p>Jupyter Notebooks (<code>.ipynb</code>) use "Kernels" to run code. A Kernel is simply your Python environment (venv or Conda).</p>
      <ul>
        <li>Create a new file and name it something like <code>lesson_1.ipynb</code>.</li>
        <li>Open the file. In the top-right corner, click on <strong>Select Kernel</strong>.</li>
        <li>Choose <strong>Python Environments</strong>.</li>
        <li>Select the exact <code>.venv</code> or Conda environment you created for this project.</li>
      </ul>
