<div class="section-card" id="toc2_">
  <h2>Selecting a Conda Environment in VS Code</h2>
  <div class="container">
    <div class="card blue">
      <h3>Interpreters</h3>
      <p>Global interpreters are system-wide; Conda interpreters are isolated for project-specific dependencies.</p>
    </div>
    <div class="card orange">
      <h3>Steps</h3>
      <ul>
        <li>Open your python project in VS Code.</li>
        <li>Click the Python interpreter version in the bottom right.</li>
        <li>Select an interpreter or enter a path (e.g., <code>/home/&lt;username&gt;/anaconda3/bin/python</code>).</li>
        <li>Choose a Conda environment from the <code>envs</code> directory.</li>
      </ul>
      <img src="../images/1-show-python-interpreter-list.png" alt="Show Python Interpreter List" style="max-width: 50vw; height: auto; display: block; margin: 30px auto; border-radius: 5px;">

  <div class="container">
    <div class="card orange">
