<div class="section-card" id="toc_venv">
  <h2>Option 1: The Fast & Lightweight Way (venv)</h2>
  <div class="container">
    <div class="card green">
      <h3>What is venv?</h3>
      <p><b>venv</b> is the standard Python tool for creating virtual environments. It is pre-installed with Python, making it the fastest way to start a project without needing external software.</p>
    </div>
    <div class="card green">
            <h3>Standard Workflow</h3>
            <p>Follow these steps to create and use an environment in your terminal:</p>
            <ul>
                <li><strong>Create:</strong> <code>python -m venv .venv</code></li>
                <li><strong>Activate (Windows):</strong> <code>.venv\Scripts\activate</code></li>
                <li><strong>Activate (Mac/Linux):</strong> <code>source .venv/bin/activate</code></li>
            </ul>
            <div class="note">
                <strong>Pro Tip:</strong> Always name your environment folder <code>.venv</code>. Modern IDEs like VS Code will detect it automatically!
            </div>
        </div>
  </div>
</div>

<div class="section-card" id="toc_conda">
  <h2>Option 2: The Professional Powerhouse (Conda)</h2>
  <div class="container">
    <!-- Conda vs Anaconda -->
    <div class="card orange">
      <h3>Conda, Miniconda, or Anaconda?</h3>
      <img src="../images/conda-anaconda-workbench.png" alt="Differences">
      <ul>
        <li><b>Conda:</b> The engine (Package & Environment Manager).</li>
        <li><b>Miniconda:</b> Lightweight (Conda + Python only). <strong>Highly Recommended!</strong></li>
        <li><b>Anaconda:</b> Heavy (Conda + 1500+ data science packages).</li>
      </ul>
    </div>
    <div class="card green">
      <h3>Installation (Miniconda)</h3>
      <p>1. Download from <a href="https://docs.anaconda.com/miniconda/" style="color: #29b6f6;">Official Site</a>.</p>
      <p>2. <b>Windows:</b> Run <code>.exe</code> and check "Add to PATH" (optional but helpful).</p>
      <p>3. <b>Mac/Linux:</b> Run <code>bash Miniconda3-latest-xxx.sh</code> in terminal.</p>
    </div>
    <div class="card pink">
      <h3>Environment Management Workflow</h3>
      <p>Managing environments allows you to isolate dependencies and maintain project-specific Python versions without conflicts.</p>
      <!-- Step 0: Conceptual Diagram -->
      <img src="../images/how-conda-works.png" style="max-width: 65%; margin-bottom: 1.5rem; opacity: 0.9;" alt="Conda Flow">
      <div style="margin-top: 1rem;">
        <ul style="list-style-type: none; padding-left: 0;">
          <!-- Step 1 -->
          <li style="margin-bottom: 1.2rem; border-bottom: 1px solid rgba(255,255,255,0.05); padding-bottom: 0.8rem;">
            <strong style="color: #f06292; font-size: 1rem;">‚ú® 1. Creation:</strong> 
            Initialize a new isolated workspace with a chosen Python version.
            <pre style="margin-top: 8px;"><code>conda create --name my_env python=3.10</code></pre>
          </li>
          <!-- Step 2 -->
          <li style="margin-bottom: 1.2rem; border-bottom: 1px solid rgba(255,255,255,0.05); padding-bottom: 0.8rem;">
            <strong style="color: #f06292; font-size: 1rem;">üöÄ 2. Activation:</strong> 
            Enter the environment to start using its specific packages.
            <pre style="margin-top: 8px;"><code>conda activate my_env</code></pre>
          </li>
          <!-- Step 3 -->
          <li style="margin-bottom: 1.2rem; border-bottom: 1px solid rgba(255,255,255,0.05); padding-bottom: 0.8rem;">
            <strong style="color: #f06292; font-size: 1rem;">üìã 3. Inventory:</strong> 
            List all environments available on your system (active one marked with *).
            <pre style="margin-top: 8px;"><code>conda env list</code></pre>
          </li>
          <!-- Step 4 -->
          <li>
            <strong style="color: #f06292; font-size: 1rem;">üßπ 4. Cleanup:</strong> 
            Remove an environment and all its packages when the project is finished.
            <pre style="margin-top: 8px;"><code>conda env remove -n my_env</code></pre>
          </li>
        </ul>
      </div>
    </div>
    <div class="card purple">
      <h3>Packages & Exporting</h3>
      <p>Conda handles complex non-Python dependencies better than pip.</p>
      <ul>
        <li><b>Install:</b> <code>conda install numpy</code></li>
        <li><b>Update:</b> <code>conda update --all</code></li>
        <li><b>Export:</b> <code>conda env export > env.yml</code> (Sharing your project)</li>
        <li><b>Import:</b> <code>conda env create -f env.yml</code></li>
      </ul>
    </div>
