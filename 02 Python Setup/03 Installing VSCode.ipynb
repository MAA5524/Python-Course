<!-- Section 1: Installing VS Code -->
<div class="section-card" id="toc1_">
  <h2>Installing VS Code</h2>
  <div class="container">
    <div class="card orange" id="toc1_1_">
      <h3>Windows Installation</h3>
      <ul>
        <li>Download the Visual Studio Code installer from the official website.</li>
        <li>Run the installer (<code>VSCodeSetup.exe</code>) and follow the setup wizard.</li>
        <li>During installation, check the "Add to PATH" option to enable launching VS Code from the terminal.</li>
        <li>Launch VS Code from the Start menu or desktop shortcut after installation.</li>
      </ul>
      <p>For more details, see <a href="https://code.visualstudio.com/Download">Install Visual Studio Code on Windows</a>.</p>
    </div>
    <div class="card orange" id="toc1_2_">
      <h3>MacOS Installation</h3>
      <ul>
        <li>Download Visual Studio Code from the official website.</li>
        <li>Open the downloaded app or extract the archive.</li>
        <li>Drag <code>Visual Studio Code.app</code> to the Applications folder.</li>
        <li>Open VS Code from the Applications folder or Launchpad.</li>
        <li>Add VS Code to the Dock via the context menu (Options > Keep in Dock).</li>
      </ul>
      <p>For more details, see <a href="https://code.visualstudio.com/Download">Install Visual Studio Code on macOS</a>.</p>
    </div>
    <div class="card orange" id="toc1_3_">
      <h3>Linux Installation</h3>
      <ul>
        <li>Install via the Ubuntu Software Center or download the <code>.deb</code> file from the official website.</li>
        <li>For command-line installation, use: <code>sudo apt install ./&lt;file&gt;.deb</code> or <code>sudo dpkg -i &lt;file&gt;.deb</code> followed by <code>sudo apt-get install -f</code> for older distributions.</li>
      </ul>
      <p>For more details, see <a href="https://code.visualstudio.com/Download">Install Visual Studio Code on Linux</a>.</p>
    </div>
  </div>
</div>

<!-- Section 2: How to Launch Visual Studio Code from Terminal -->
<div class="section-card" id="toc2_">
  <h2>How to Launch Visual Studio Code from Terminal</h2>
  <div class="container">
    <div class="card purple">
      <h3>Steps to Launch</h3>
      <ul>
        <li>Ensure VS Code is installed.</li>
        <li>Add VS Code to the system PATH if needed.</li>
        <li>Open a terminal and type <code>code .</code> to open the current directory as a workspace.</li>
      </ul>
    </div>
    <div class="card purple">
      <h3>Difference: <code>code</code> vs <code>code .</code></h3>
      <p><code>code</code> opens VS Code with a blank workspace; <code>code .</code> opens the current directory as the workspace.</p>
    </div>
    <div class="card purple">
      <h3>Workspace Benefits</h3>
      <p>Opening a directory as a workspace groups related files, enabling project-specific settings and efficient file management.</p>
    </div>
  </div>
</div>
