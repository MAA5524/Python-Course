<!-- Section 1: Understanding the Differences: Conda, Miniconda, and Anaconda -->
<div class="section-card" id="toc1_">
  <h2>Understanding the Differences: Conda, Miniconda, and Anaconda</h2>
  <div class="container">
    <div class="card image-card orange">
      <h3>Overview</h3>
      <img src="../images/conda-anaconda-workbench.png" alt="Conda Differences" style="max-width: 50vw; height: auto; display: block; margin: 10px auto; border-radius: 5px;">
    </div>
    <div class="card orange">
      <h3>Conda</h3>
      <p>Conda is the package manager at the core of Miniconda and Anaconda. It installs, updates, and manages packages and dependencies, supporting multiple platforms and languages.</p>
    </div>
    <div class="card orange">
      <h3>Miniconda</h3>
      <p>Miniconda is a minimal version of Anaconda, including only Conda and essential dependencies for a lightweight, customizable installation.</p>
    </div>
    <div class="card orange">
      <h3>Anaconda</h3>
      <p>Anaconda is a distribution including Conda and pre-installed Python packages, data science libraries, and tools, tailored for scientific computing and machine learning.</p>
    </div>
  </div>
</div>

<!-- Section 2: Why Prefer Miniconda -->
<div class="section-card" id="toc2_">
  <h2>Why Prefer Miniconda</h2>
  <div class="container">
    <div class="card purple">
      <h3>Customizability</h3>
      <p>Minicondaâ€™s minimalistic approach allows installing only needed packages, ensuring lightweight and efficient environments.</p>
    </div>
    <div class="card purple">
      <h3>Reduced Disk Space</h3>
      <p>Miniconda consumes less disk space than Anaconda, ideal for systems with limited storage or multiple environments.</p>
    </div>
    <div class="card purple">
      <h3>Faster Installation</h3>
      <p>Miniconda has a smaller download size and faster installation process, convenient for quick setups.</p>
    </div>
  </div>
</div>

<!-- Section 3: Installing miniconda -->
<div class="section-card" id="toc3_">
  <h2>Installing miniconda</h2>
  <div class="container">
    <div class="card image-card green">
      <h3>Overview</h3>
      <img src="../images/miniconda.svg" alt="Miniconda Installation" style="max-width: 30vw; height: auto; display: block; margin: 10px auto; border-radius: 5px;">
    </div>

    </div>
    <div class="card green" id="toc3_3_">
      <h3>macOS Installation</h3>
      <p>Download the installer script, run <code>bash Miniconda3-latest-MacOSX-x86_64.sh</code>, follow prompts, and add to PATH by restarting the terminal.</p>
    </div>
  </div>
</div>

<!-- Section 4: Conda Environments -->
<div class="section-card" id="toc4_">
  <h2>Conda Environments</h2>
  <div class="container">
    <div class="card image-card pink">
      <h3>Overview</h3>
      <img src="../images/how-conda-works.png" alt="How Conda Works" style="max-width: 50vw; height: auto; display: block; margin: 10px auto; border-radius: 5px;">
    </div>
    <div class="card pink" id="toc4_1_">
      <h3>Creating a New Environment</h3>
      <p>Use <code>conda create -n my_env python=3.8</code> to create an environment with a specific Python version.</p>
    </div>
    <div class="card pink" id="toc4_2_">
      <h3>Activating and Deactivating Environments</h3>
      <p>Use <code>conda activate my_env</code> to activate and <code>conda deactivate</code> to deactivate an environment.</p>
      <img src="../images/conda-activate.jpeg" alt="Conda Activate" style="max-width: 50vw; height: auto; display: block; margin: 10px auto; border-radius: 5px;">
    </div>
    <div class="card pink" id="toc4_3_">
      <h3>Listing Environments and Packages</h3>
      <p>Use <code>conda env list</code> for environments and <code>conda list</code> for packages in the active environment.</p>
    </div>
    <div class="card pink" id="toc4_4_">
      <h3>Removing Environments</h3>
      <p>Use <code>conda env remove -n my_env</code> to remove an environment.</p>
    </div>
  </div>
</div>

<!-- Section 5: Installing and Managing Packages -->
<div class="section-card" id="toc5_">
  <h2>Installing and Managing Packages</h2>
  <div class="container">
    <div class="card orange" id="toc5_1_">
      <h3>Installing Packages</h3>
      <p>Use <code>conda install PACKAGE_NAME</code> or <code>conda install PACKAGE_NAME=VERSION</code> to install packages or specific versions.</p>
    </div>
    <div class="card orange" id="toc5_1_1_">
      <h3>Installing Packages with pip</h3>
      <p>Use <code>pip install PACKAGE_NAME</code> for packages not available through Conda, though Conda is preferred for dependencies.</p>
    </div>
    <div class="card orange" id="toc5_2_">
      <h3>Updating Packages</h3>
      <p>Use <code>conda update PACKAGE_NAME</code> for a package or <code>conda update --all</code> for all packages.</p>
    </div>
    <div class="card orange" id="toc5_3_">
      <h3>Removing Packages</h3>
      <p>Use <code>conda remove PACKAGE_NAME</code> to remove a package.</p>
    </div>
  </div>
</div>

<!-- Section 6: Environment Files -->
<div class="section-card" id="toc6_">
  <h2>Environment Files</h2>
  <div class="container">
    <div class="card red" id="toc6_1_">
      <h3>Exporting an Environment</h3>
      <p>Use <code>conda env export > environment.yml</code> to export the environment configuration.</p>
    </div>
    <div class="card red" id="toc6_2_">
      <h3>Creating an Environment from a File</h3>
      <p>Use <code>conda env create -f environment.yml</code> to create an environment from a file.</p>
    </div>
  </div>
</div>
