<!-- Section 1: What Is a Package Manager -->
<div class="section-card" id="toc1_">
    <h2>What Is a Package Manager</h2>
    <div class="container">
        <div class="card image-card orange">
            <h3>Overview</h3>
            <p>A package manager is like a store manager who finds, organizes, and delivers the tools (libraries) you need for your Python projects, saving you time and effort.</p>
            <img src="../images/packege.jpg" alt="Package Manager Overview" style="max-width: 50vw; height: auto; display: block; margin: 10px auto; border-radius: 5px;">
        </div>
        <div class="card orange">
            <h3>What Are Packages?</h3>
            <p>Packages are pre-built libraries that add functionality to your code, like NumPy for math or Django for web apps. Without a package manager, installing them manually is slow and error-prone.</p>
            <p>Example: To use Pandas for data analysis, you can install it with one command:</p>
            <pre><code>pip install pandas</code></pre>
        </div>
        <div class="card orange">
            <h3>Key Functions</h3>
            <ul>
                <li><strong>Install Libraries:</strong> Downloads and sets up libraries with a single command.</li>
                <li><strong>Handle Dependencies:</strong> Automatically installs other required libraries (e.g., NumPy for Pandas).</li>
                <li><strong>Update Packages:</strong> Keeps libraries up-to-date with new versions.</li>
                <li><strong>Remove Packages:</strong> Cleans up unused libraries to save space.</li>
            </ul>
        </div>
        <div class="card orange">
            <h3>Popular Package Managers</h3>
            <p>Examples include:</p>
            <ul>
                <li><strong>pip:</strong> Pythonâ€™s default package manager, perfect for installing libraries from PyPI.</li>
                <li><strong>Conda:</strong> A powerful manager for Python and non-Python libraries, ideal for data science.</li>
                <li><strong>npm:</strong> Used for JavaScript, showing package managers exist in other languages too.</li>
            </ul>
        </div>
    </div>
</div>

<!-- Section 2: How Package Managers Work (Previously Provided) -->
<div class="section-card" id="toc2_">
    <h2>How Package Managers Work</h2>
    <div class="container">
        <div class="card image-card purple">
            <h3>Overview</h3>
            <p>A package manager acts like a personal assistant for your coding projects, finding, installing, and organizing the tools (libraries) you need from online repositories.</p>
            <img src="../images/how-package-manager-work.png" alt="Package Manager Workflow" style="max-width: 50vw; height: auto; display: block; margin: 10px auto; border-radius: 5px;">
        </div>
        <div class="card purple">
            <h3>Steps of Package Management</h3>
            <p>Package managers follow a clear process to manage libraries:</p>
            <ul>
                <li><strong>Search:</strong> They search repositories (like PyPI or Conda Forge) to find the requested package.</li>
                <li><strong>Download:</strong> They download the package and its dependencies.</li>
                <li><strong>Install:</strong> They install the package, ensuring compatibility with your system.</li>
                <li><strong>Manage Dependencies:</strong> They resolve conflicts by selecting compatible versions.</li>
            </ul>
            <p>Example: To install Pandas:</p>
            <pre><code>pip install pandas</code></pre>
        </div>
        <div class="card purple">
            <h3>Virtual Environments</h3>
            <p>Package managers create virtual environments to keep projects separate, preventing conflicts between different package versions.</p>
            <p>Example with Conda:</p>
            <pre><code>conda create -n my_project python=3.9
conda activate my_project
conda install pandas</code></pre>
        </div>
        <div class="card purple">
            <h3>Real-World Example</h3>
            <p>For a data analysis app for Digikala, a package manager:</p>
            <ul>
                <li>Finds Pandas in PyPI or Conda Forge.</li>
                <li>Downloads Pandas and its dependencies (e.g., NumPy).</li>
                <li>Installs them in an isolated environment.</li>
            </ul>
            <pre><code>import pandas as pd
data = pd.read_csv("sales_data.csv")
print(data.head())</code></pre>
        </div>
    </div>
</div>

<!-- Section 3: Benefits of Using a Package Manager -->
<div class="section-card" id="toc3_">
    <h2>Benefits of Using a Package Manager</h2>
    <div class="container">
        <div class="card image-card green">
            <h3>Why Use a Package Manager?</h3>
            <p>Package managers simplify coding by automating library management, letting you focus on building projects like a web app or data analysis tool.</p>
            <img src="https://geekflare.com/wp-content/uploads/2023/09/7-package-manager-functions.png" alt="Benefits of Package Manager" style="max-width: 50vw; height: auto; display: block; margin: 10px auto; border-radius: 5px;">
        </div>
        <div class="card green">
            <h3>Key Benefits</h3>
            <ul>
                <li><strong>Saves Time:</strong> Installs libraries quickly with one command, like <code>pip install django</code> for a web app.</li>
                <li><strong>Prevents Conflicts:</strong> Uses virtual environments to keep project tools separate, avoiding version mismatches.</li>
                <li><strong>Ensures Consistency:</strong> Records exact package versions, so your project works the same on any system.</li>
                <li><strong>Simplifies Updates:</strong> Easily updates libraries to their latest versions with commands like <code>pip install --upgrade pandas</code>.</li>
            </ul>
        </div>
        <div class="card green">
            <h3>Example in Action</h3>
            <p>For a machine learning project, you need TensorFlow. A package manager like Conda:</p>
            <ul>
                <li>Installs TensorFlow and its dependencies: <code>conda install tensorflow</code></li>
                <li>Creates an isolated environment to avoid conflicts with other projects.</li>
                <li>Ensures the project runs consistently on any computer.</li>
            </ul>
            <pre><code>import tensorflow as tf
model = tf.keras.Sequential()
print("TensorFlow is ready!")</code></pre>
        </div>
        <div class="card green">
            <h3>Why Conda Shines</h3>
            <p>Conda is great for complex projects like data science because it handles non-Python dependencies (e.g., C libraries) and provides pre-compiled packages for fast setup.</p>
        </div>
    </div>
</div>

<!-- Section 4: Package Repository -->
<div class="section-card" id="toc4_">
    <h2>Package Repository</h2>
    <div class="container">
        <div class="card image-card orange">
            <h3>What Is a Package Repository?</h3>
            <p>A package repository is like an online library where developers share and download code packages, making it easy to access tools for your projects.</p>
            <img src="../images/pypi-pip.png" alt="Package Repository" style="max-width: 50vw; height: auto; display: block; margin: 10px auto; border-radius: 5px;">
        </div>
        <div class="card orange">
            <h3>Role of Repositories</h3>
            <p>Repositories store thousands of packages, allowing package managers to find and install them. They also keep packages updated and secure.</p>
            <p>Example: Installing Flask for a web app:</p>
            <pre><code>pip install flask</code></pre>
            <p>This command fetches Flask from PyPI, the largest Python repository.</p>
        </div>
        <div class="card orange">
            <h3>Popular Repositories</h3>
            <ul>
                <li><strong>PyPI:</strong> The Python Package Index, hosting over 400,000 Python packages like Pandas and Django.</li>
                <li><strong>Conda Forge:</strong> A community-driven repository with pre-compiled packages for Python and non-Python tools, ideal for data science.</li>
            </ul>
        </div>
        <div class="card orange">
            <h3>Real-World Use</h3>
            <p>For a project like building a website similar to Aparat, you can use PyPI to install Django:</p>
            <pre><code>pip install django</code></pre>
            <p>Or use Conda Forge for a data science project:</p>
            <pre><code>conda install -c conda-forge scikit-learn</code></pre>
            <div class="note">
                <strong>Note:</strong> Repositories like PyPI are like a digital marketplace where you can find tools for any Python project, from web apps to AI.
            </div>
        </div>
    </div>
</div>

<!-- Section 5: Conda vs. Pip -->
<div class="section-card" id="toc5_">
    <h2>Conda vs. Pip</h2>
    <div class="container">
        <div class="card image-card red">
            <p>Pip and Conda are popular package managers, but they serve different needs. Pip is great for Python libraries, while Conda excels in complex projects with non-Python dependencies.</p>
            <img src="https://miro.medium.com/v2/resize:fit:1200/1*cD6_7-slJ-fqo1evJTQbMw.png" alt="Conda vs Pip" style="max-width: 60vw; height: auto; display: block; margin: 10px auto; border-radius: 5px;">
        </div>
        <div class="card red">
            <h3>Key Differences</h3>
            <table>
                <tr>
                    <th>Feature</th>
                    <th>Pip</th>
                    <th>Conda</th>
                </tr>
                <tr>
                    <td><strong>Purpose</strong></td>
                    <td>Installs Python packages from PyPI</td>
                    <td>Manages Python and non-Python packages, plus environments</td>
                </tr>
                <tr>
                    <td><strong>Virtual Environments</strong></td>
                    <td>Requires tools like virtualenv for environments</td>
                    <td>Built-in environment management</td>
                </tr>
                <tr>
                    <td><strong>Binary Packages</strong></td>
                    <td>Some packages need compilation</td>
                    <td>Provides pre-compiled binaries</td>
                </tr>
                <tr>
                    <td><strong>Dependency Management</strong></td>
                    <td>Basic dependency resolution</td>
                    <td>Advanced resolution for complex dependencies</td>
                </tr>
                <tr>
                    <td><strong>Use Case</strong></td>
                    <td>Ideal for simple Python projects (e.g., web apps)</td>
                    <td>Best for data science and scientific computing</td>
                </tr>
            </table>
        </div>
        <div class="card red">
            <h3>Example Comparison</h3>
            <p>For a web app like Aparat, use pip to install Django:</p>
            <pre><code>pip install django</code></pre>
            <p>For a data science project, use Conda to manage complex libraries like TensorFlow:</p>
            <pre><code>conda create -n ml_project python=3.8
conda install tensorflow</code></pre>
            <div class="note">
                <strong>Note:</strong> Choose pip for lightweight Python projects and Conda for projects needing robust environment and dependency management.
            </div>
        </div>
    </div>
</div>

<!-- Section 6: PyPI vs. Conda Forge -->
<div class="section-card" id="toc6_">
    <h2>PyPI vs. Conda Forge</h2>
    <div class="container">
        <div class="card image-card green">
            <p>PyPI and Conda Forge are repositories where package managers find libraries. PyPI is the largest Python repository, while Conda Forge focuses on pre-compiled packages for complex projects.</p>
        </div>
        <div class="card green">
            <h3>Key Differences</h3>
            <table>
                <tr>
                    <th>Feature</th>
                    <th>PyPI</th>
                    <th>Conda Forge</th>
                </tr>
                <tr>
                    <td><strong>Package Types</strong></td>
                    <td>Python packages only</td>
                    <td>Python and non-Python packages</td>
                </tr>
                <tr>
                    <td><strong>Binary Packages</strong></td>
                    <td>Not always pre-compiled</td>
                    <td>Emphasizes pre-compiled binaries</td>
                </tr>
                <tr>
                    <td><strong>Build System</strong></td>
                    <td>Depends on package maintainers</td>
                    <td>Uses standardized Conda Build</td>
                </tr>
                <tr>
                    <td><strong>Community</strong></td>
                    <td>Broad Python community</td>
                    <td>Focused data science community</td>
                </tr>
                <tr>
                    <td><strong>Use Case</strong></td>
                    <td>General Python projects (e.g., web apps)</td>
                    <td>Scientific and data-intensive projects</td>
                </tr>
            </table>
        </div>
        <div class="card green">
            <h3>Example Use Cases</h3>
            <p>For a web app like Digikala, use PyPI with pip:</p>
            <pre><code>pip install django</code></pre>
            <p>For an AI project, use Conda Forge with Conda:</p>
            <pre><code>conda install -c conda-forge pytorch</code></pre>
            <div class="note">
                <strong>Note:</strong> PyPI is great for general Python libraries, while Conda Forge is ideal for projects needing reliable, pre-compiled packages.
            </div>
        </div>
    </div>
</div>
