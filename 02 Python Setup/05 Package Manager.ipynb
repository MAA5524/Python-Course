<!-- Section 1: What Is a Package Manager -->
<div class="section-card" id="toc1_">
  <h2>What Is a Package Manager</h2>
  <div class="container">
    <div class="card image-card green">
      <h3>Overview</h3>
      <img src="../images/packege.jpg" alt="Package Manager" style="max-width: 50vw; height: auto; display: block; margin: 10px auto; border-radius: 5px;">
    </div>
    <div class="card blue">
      <h3>Definition</h3>
      <p>A package manager is a tool that automates installing, updating, configuring, and removing software packages, acting like a librarian to manage code libraries and tools.</p>
    </div>
    <div class="card orange">
      <h3>Packages</h3>
      <p>Packages are bundles of code (e.g., NumPy for math, Pandas for data analysis) that add functionality to projects. Manual installation is tedious and error-prone.</p>
    </div>
    <div class="card purple">
      <h3>Functionality</h3>
      <ul>
        <li><strong>Simplifying Installation:</strong> One command fetches and installs everything.</li>
        <li><strong>Managing Dependencies:</strong> Ensures all required packages are compatible.</li>
        <li><strong>Enabling Updates:</strong> Keeps tools current.</li>
        <li><strong>Supporting Environments:</strong> Isolates projects to prevent interference.</li>
      </ul>
    </div>
    <div class="card red">
      <h3>Examples</h3>
      <p>Examples include pip (for Python), npm (for JavaScript), and conda (a versatile manager).</p>
    </div>
  </div>
</div>

  <div class="container">
    <div class="card image-card green">
      <h3>Overview</h3>
      <img src="https://geekflare.com/wp-content/uploads/2023/09/7-package-manager-functions.png" alt="How Package Managers Work" style="max-width: 50vw; height: auto; display: block; margin: 10px auto; border-radius: 5px;">

    <div class="card blue">

      <p>Packages are external libraries or tools (e.g., for data handling or plotting) that extend Pythonâ€™s capabilities, saving time with pre-built functionality.</p>

<!-- Section 3: Benefits of Using a Package Manager -->
<div class="section-card" id="toc3_">
  <h2>Benefits of Using a Package Manager</h2>
  <div class="container">
    <div class="card image-card green">
      <h3>Overview</h3>
      <img src="../images/conda-python.jpg" alt="Benefits of Package Manager" style="max-width: 50vw; height: auto; display: block; margin: 10px auto; border-radius: 5px;">
    </div>
    <div class="card blue">
      <h3>Benefits</h3>
      <ul>

<!-- Section 5: Conda vs. Pip -->
<div class="section-card" id="toc5_">
  <h2>Conda vs. Pip</h2>
  <div class="container">
    <div class="card blue">
      <h3>Overview</h3>
      <p>Conda is a cross-platform package manager for pre-built packages, while pip is Pythonâ€™s default manager for installing from PyPI.</p>
    </div>
    <div class="card orange">
      <h3>Comparison Table</h3>
      <table>
        <tr>
          <th></th>
          <th>Conda</th>
          <th>pip</th>
        </tr>
        <tr>
          <td><strong>Installation</strong></td>
          <td>Installs packages and manages environments simultaneously</td>
          <td>Installs packages only</td>
        </tr>
        <tr>
          <td><strong>Environment Management</strong></td>
          <td>Manages isolated environments with specific package versions</td>
          <td>Does not handle environment management</td>
        </tr>
        <tr>
          <td><strong>Binary Packages</strong></td>
          <td>Provides pre-compiled binary packages for easy installation</td>
          <td>Requires package compilation in some cases</td>
        </tr>
        <tr>
          <td><strong>Platform Compatibility</strong></td>
          <td>Ensures package compatibility across different operating systems</td>
          <td>Relies on platform-specific package distributions</td>
        </tr>
        <tr>
          <td><strong>Package Availability</strong></td>
          <td>Offers a wide range of packages, including non-Python dependencies</td>
          <td>Focuses primarily on Python packages</td>
        </tr>
        <tr>
          <td><strong>Package Channels</strong></td>
          <td>Supports Conda channels for package distribution and versioning</td>
          <td>Relies on PyPI as the default package repository</td>
        </tr>
        <tr>
          <td><strong>Dependency Resolution</strong></td>
          <td>Solves complex dependency chains to ensure compatibility</td>
          <td>Relies on package maintainers to provide dependency information</td>
        </tr>
      </table>
    </div>
  </div>
</div>

<!-- Section 6: PyPI vs. Conda Forge -->
<div class="section-card" id="toc6_">
  <h2>PyPI vs. Conda Forge</h2>
  <div class="container">
    <div class="card blue">
      <h3>PyPI</h3>
      <p>PyPI is the largest repository of Python packages, hosting an extensive collection contributed by developers worldwide, supporting binary distribution but not guaranteed for all packages.</p>
    </div>
    <div class="card orange">
      <h3>Conda Forge</h3>
      <p>Conda Forge is a community-driven platform emphasizing pre-compiled binary packages for Python and non-Python dependencies, ensuring a robust environment.</p>
    </div>
    <div class="card purple">
      <h3>Comparison Table</h3>
      <table>
        <tr>
          <th></th>
          <th>PyPI</th>
          <th>Conda Forge</th>
        </tr>
        <tr>
          <td><strong>Package Availability</strong></td>
          <td>Largest repository of Python packages</td>
          <td>Comprehensive collection of Python and non-Python packages</td>
        </tr>
        <tr>
          <td><strong>Binary Packages</strong></td>
          <td>Supports binary distribution but not guaranteed for all packages</td>
          <td>Emphasizes pre-compiled binary packages for various platforms</td>
        </tr>
        <tr>
          <td><strong>Package Versioning</strong></td>
          <td>Package maintainers handle versioning and releases</td>
          <td>Community-driven versioning and continuous integration process</td>
        </tr>
        <tr>
          <td><strong>Dependency Management</strong></td>
          <td>Relies on package maintainers to define dependencies</td>
          <td>Centralized dependency management through Conda environment files</td>
        </tr>
        <tr>
          <td><strong>Build System</strong></td>
          <td>Relies on package maintainers to provide build scripts</td>
          <td>Utilizes the Conda build system for consistent package building</td>
        </tr>
      </table>
    </div>
  </div>
</div>
