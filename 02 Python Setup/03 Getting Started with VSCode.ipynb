<div class="section-card" id="toc1_">
  <h2>Opening VS Code in Different Directories and Files</h2>
  <div class="container">
    <div class="card orange">
      <h3>Opening Directories</h3>
      <ul>
        <li>Use <code>File -> Open Folder...</code> or <code>Ctrl+K Ctrl+O</code> (Windows/Linux) / <code>Cmd+K Cmd+O</code> (Mac) to browse and open a folder.</li>
        <li>Right-click a folder in file explorer and select "Open with Code."</li>
        <li>In a terminal, navigate to the directory with <code>cd</code> and type <code>code .</code> to open it in VS Code.</li>
      </ul>
    </div>
    <div class="card orange">
      <h3>Opening Single Files</h3>
      <ul>
        <li>Use <code>File -> Open...</code> or <code>Ctrl+O</code> (Windows/Linux) / <code>Cmd+O</code> (Mac) to browse and open a file.</li>
        <li>Right-click a file in file explorer and select "Open with Code."</li>
      </ul>
    </div>
  </div>
</div>
